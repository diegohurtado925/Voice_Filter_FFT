## Experimento 1: Filtro de Banda "Duro" con FFT Global
**Hipótesis:** Asilar la voz eliminando todas las frecuencias que estén fuera del rango típico de la voz humana (hacer cero las frecuencias menores a 300 Hz y mayores a 3000 Hz) aplicadas a la totalidad del arreglo.


```python
import numpy as np
from scipy.io import wavfile
from scipy.fft import fft, ifft, fftfreq
from IPython.display import Audio

# Leer el archivo de audio original con ruido
fs, data = wavfile.read('Audio1.wav')

# Normalizar el arreglo de la señal de entrada
if len(data.shape) > 1:
    data = data[:, 0]  # Convertir a mono si es necesario
data = data / np.max(np.abs(data))

print(f"Frecuencia de muestreo (fs): {fs} Hz")
print(f"Total de valores en el arreglo: {len(data)}")
```

```python
# 1. Aplicar la Transformada Rápida de Fourier (FFT) a TODO el arreglo
N = len(data)
senal_fft = fft(data)
frecuencias = fftfreq(N, 1/fs)

# 2. Crear una máscara de frecuencias (Mantener solo entre 300 Hz y 3000 Hz)
mascara_banda = (np.abs(frecuencias) >= 300) & (np.abs(frecuencias) <= 3000)

# 3. Aplicar el hachazo: Hacer cero todo lo que esté fuera de ese rango
senal_fft_filtrada = senal_fft * mascara_banda

# 4. Reconstruir la señal con la Transformada Inversa (IFFT)
audio_exp1 = np.real(ifft(senal_fft_filtrada))

# Exportar el audio fallido para demostración empírica
wavfile.write('Audio2.wav', fs, audio_exp1.astype(np.float32))

# Reproductor
Audio(audio_exp1, rate=fs)
```

**Resultado** Al aplicar la FFT a todo el archivo de audio como si fuera un solo bloque estático, la eliminación drástica de componentes de frecuencia mutilará los armónicos naturales de la voz, haciéndola sonar ahogada y robótica. Además, el ruido del ventilador que comparte frecuencias con la voz (entre 300 y 3000 Hz) permanecerá intacto.
